3. Un reservorio de longitud L tiene una seccion transveral en forma de semicirculo con radio r.

\begin{equation}
V=L \left ( \frac{1}{2}\pi r^2 -r^2 \arcsin{\frac{h}{r}}-h(r^2-h^2) ^{1/2} \right)
\end{equation}

Determinar la profundidad del agua para L=10pies, r=1pies, V=12.4pies


![Reservorio](https://raw.githubusercontent.com/hernansalinas/Metodos_Computacionales_taller/main/sesiones/images/ecuacion_1_variable/Fig1.png)

```python
#Constantes en pies.
L = 10
r = 1
V = 12.4



g = lambda h : V-L*((0.5*np.pi*r**2)-(np.arcsin(h/r)*r**2)-(h*(r**2-h**2)**0.5)) #Función dependiente de h.
h = np.linspace(0, 1) #Dominio de  h.

#Graficar g(h).

plt.figure()
plt.title("g(h)",fontdict={'family': 'serif', 'color' : 'darkblue','weight': 'bold','size': 18})
plt.plot(h, g(h),"red",label=("g(h)"))
plt.ylabel("g(h)")
plt.xlabel("h")
plt.legend()
plt.grid()
plt.show()

#Para encontrar el h que cumple para L = 10, r = 1, V = 12.4 para esto ese necesario encontrar las raíces de la función g(h).

#Metodo de bisección

rootb = optimize.bisect(g, 0.0, 0.2)

#Metodo Newton Raphson

rootNR = optimize.newton(g, 0.2)

print("la profundida h para las condicones L = 10, r = 1, V = 12.4  utilizando el metodo de bisección es:{} y utilizando Newton Raphson es:{}\nDfieren por: {}" .format(rootb,rootNR,np.absolute(rootb-rootNR)))

#Teniendo h, ahora se debe calcular la "altura del agua".

ab = 1 - rootb #Usando la raíz calculada por bisección.
aNR = 1 - rootNR #Usando la raíz calculada por Newton Raphson.

print("\n\nla \"altura del agua\" utilizando la raíz calculada por bisección es: {} pies y utilizando las raíz Newton Raphson es: {} pies.\nDifieren por: {}".format(ab,aNR,np.absolute(ab-aNR)))

```

4.Resolver la siguiente situación

En el diseño de los vehiculos para todo tipo de terreno,  es necesario tener encuentra fallas cuando se trata de liberar dos tipos de obstáculos. Una es la falla por rozamiento, y ocurre cuando el vehiculo 
intenta cruzar un obstáculo que hace que su fondo toque el suelo. La otra recibie el nombre de falla por colision de la defensa delantera y ocurre cuando el vehiculo desciende por una zanja y la defensa delantera toca el suelo.



La figura muestra las componentes asociados al segundo tipo  de falla. En ella se indeca el ángulo máximo $\alpha$ que puede alcanzar un vehiculo cuando $\beta$ es el angulo maximo en que no ocurre la falla 
por rozamiento y satisface la siguiente ecuación: 


\begin{equation}
A \sin(\alpha) \cos(\alpha)+B \sin^2(\alpha) - C \cos(\alpha) -  E  \sin(\alpha)=0
\end{equation}

donde; 
\begin{equation}
A = l \sin(\beta)
\end{equation}

\begin{equation}
B = l \cos(\beta)
\end{equation}
\begin{equation}
C = (h + 0.5 D )\sin(\beta) - 0.5 D \tan(\beta)
\end{equation}

\begin{equation}
E = (h + 0.5 D ) \cos(\beta) - 0.5 * D
\end{equation}


Para $l=89$pulg, $h=49$pul, $D=55$, $\beta=11.5$ , el angulo se aproximara a 33. Verificar este resultados


![fallas_vehiculo](https://raw.githubusercontent.com/hernansalinas/Metodos_Computacionales_taller/main/sesiones/images/ecuacion_1_variable/Fig2.png)

```python
#Definir A,B,C,E.

l = 89 #Pulgadas
h = 49 #Pulgadas
D = 55 #Pulgadas
beta = np.deg2rad(11.5) #Pasar de grados a radianes.

A = l*np.sin(beta)
B = l*np.cos(beta)
C = (h + 0.5*D)*np.sin(beta)-0.5*D*np.tan(beta)
E = (h + 0.5*D)*np.cos(beta) - 0.5*D 

#Definir la función dependiente de alpha

f = lambda alpha : A*np.sin(alpha)*np.cos(alpha) + B*np.sin(alpha)**2 - C*np.cos(alpha) - E*np.sin(alpha)
alpha =  np.linspace(0, 2*np.pi) #Dominio para alpha


plt.figure()
plt.title("g("r"$\alpha$"")",fontdict={'family': 'serif', 'color' : 'darkblue','weight': 'bold','size': 18})
plt.plot(alpha, f(alpha),"red",label=("g("r"$\alpha$"")"))
plt.ylabel("g("r"$\alpha$"")")
plt.xlabel(r"$\alpha$")
plt.legend()
plt.grid()
plt.show()

#Como se ve en la grafica g(alpha) tiene varias raices en [0,2pi], tomar la primera que esta en [0,1rad]. 

#Metodo de bisección

rootb = optimize.bisect(f, 0.1, 1)
rootb_grados = np.rad2deg(rootb)

#Metodo Newton Raphson

rootNR = optimize.newton(f, 0.5)
rootNR_grados = np.rad2deg(rootNR)

print("El valor del angulo \u03B1 utilizando el metodo de bisección es: {} radianes o {} grados y utilizando Newton Rapshon es: {} radianes o {} grados\n\nLos dos metodos difieren por: {} radianes o {} grados".format(rootb,rootb_grados,rootNR,rootNR_grados,np.absolute(rootb - rootNR),np.absolute(rootb_grados - rootNR_grados)))

```

5. Construir un fractal empleando el metodo de Newton-Raphson, 
(Reproducir código y entender)
\begin{equation}
z_{n+1} = z_n -a \frac{f(z)}{f'(z)} 
\end{equation}



\begin{equation}
f'(z) = \frac{ f(z+h)+f(z) }{h}
\end{equation}

```python
import numpy as np
import matplotlib.pylab as plt
from PIL import Image
```

```python
Npx_X = 80 #Numero de pixeles en x
Npx_Y = 80 #Numero de pixeles en y

image = Image.new("RGB", (Npx_X, Npx_Y)) #Creacion de la imagen inicial

npx = np.arange(0, Npx_X)
npy = np.arange(0, Npx_Y)

xmin = -1
xmax = 1
ymin = -1
ymax = 1
#Transformación al plano complejo en un rectangulo (xmin, xmax), (ymin, ymax)
zx = npx * (xmax - xmin) / (Npx_X - 1)   + xmin
zy = npy * (ymax-ymin) / (Npx_Y - 1) + ymin
```

```python
#Definicion  la funcion y las derivada 
def f(z): 
  return z**3 - 1

def fp(z): 
  return 3*z**2

h = 1E-4
def fpn(z, h): 
  return (f(z+h)-f(z))/h
  
```

```python
a = 0.2
Niteraciones = 120
eps = 1E-8

x = - 1
for zx_ in zx:
  x = x + 1 
  y = -1
  for zy_ in zy: 
      y = y + 1
      Z_0 = complex(zx_, zy_)
    
      for i in range(0, Niteraciones): 
        Z = Z_0 - a * f(Z_0)/fp(Z_0)

        if( abs(Z-Z_0)<= eps):
          break 
          
        Z_0 = Z
      #print(i)
      #print(i)
      r, g, b = i % 4 *64 , i % 8 *16, i% 16* 16
      
      image.putpixel((x,y), (r,g,b) )
```
