## Cancelación sustractiva



**serieSn**. Un ejemplo de la cancelacion substractiva es en la suma de series alternantes. Considere la serie $\mathrm{S}_N$,  escrita  matematicamente de tres formas diferentes:

- 
\begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1} 
\end{equation}

- Sumando números pares e impares.

\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

- Combinando las series de una forma analítica

\begin{equation}
\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}
\end{equation}



Las tres sumas $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$ son matemáticamente iguales, pero los resultados  pueden tener resultados numéricos diferentes.


- Escribir un programa que realice el cálculo de $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$

##**Programa que realiza el cálculo de $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$**

````python
# Primera serie
def serie1 (n):
  """
  input
  ------
  n : hasta que valor va la serie
  ------
  output
  ------
  num : retorna la suma de la serie
  """
  N = 2*n #Se multiplica por 2 dado que la serie va hasta (2N)
  sum = 0
  for i in range (1,N+1):
    num = ((((-1)**i)*(i))/(i+1))
    sum = sum + num
  return sum

#Segunda serie
def serie2 (n):
  """
  input
  ------
  n : hasta que valor va la serie
  ------
  output
  ------
  num : retorna la suma de la serie
  """
  sum1 = 0
  sum2 = 0
  for i in range (1,n+1):
    num1 = (((2*i)-1)/(2*i))
    sum1 = sum1 + num1
  for i in range (1,n+1):  
    num2 = (2*i/((2*i)+1))
    sum2 = sum2 + num2
  return sum2 - sum1

#Tercera serie 
def serie3 (n):
  """
  input
  ------
  n : hasta que valor va la serie
  ------
  output
  ------
  num : retorna la suma de la serie
  """
  sum = 0
  for i in range (1,n+1):
    num = ((1)/((2*i)*((2*i)+1)))
    sum = sum + num
  return sum
      
````

````python
print ('El valor de la primera serie es:  ',serie1( n = 10 ))
print ('El valor de la segunda serie es:  ',serie2( n = 10 ))
print ('El valor de la tercera serie es:  ',serie3( n = 10 ))
````

##**Estudio de la Convergencia**

````python
n = 10  # Número hasta el cual va la serie

# Serie 1
conv_s1 = []
for i in range (1, n+1):
  conv = serie1 (i)
  conv_s1.append(conv)
print ('Convergencia Serie 1: \n',conv_s1) 
# Serie 2
conv_s2 = []
for i in range (1, n+1):
  conv = serie2 (i)
  conv_s2.append(conv)
print ('Convergencia Serie 2: \n',conv_s2)  

# Serie 3
conv_s3 = []
for i in range (1, n+1):
  conv = serie3 (i)
  conv_s3.append(conv)
print ('Convergencia Serie 3: \n',conv_s3)  
````

## **Errores de redondeo**. Caos numérico en un mapa logístico y errores de punto flotante (Errores de redondeo)

> Un ejemplo clásico de caos,  es el comportamiento no lineal en las interaciones de un mapa logistico 

\begin{equation}
x_{n+1}=f(x_n)=rx_n(1-x_n)
\end{equation}

> con $x\in (0,1)$ y $r\in(0,4)$ se pueden producir varios comportamientos sorprendentes.


> ### Problema: 
Encontrar tres formas diferentes de expresar $f(x)$  y calcular la evolución de la misma condición inicial después de cientos de iteraciones. Para este problema, será extremadamente útil  ver sus resultados gráficamente; construya listas de números y llame la libreria matplotlib. 


>#### Hint: valores numéricos de r y x0 que puede tomar:


```python

       ```
       

##**Las tres maneras que tomo para escribir $f(x_n)=rx_n(1-x_n)$ es:**

-  $f(x_n)=rx_n-rx_n^2$
-  $f(x_n)=r(x_n-x_n^2)$
-  $f(x_n)=x_n^2(r/x_n-1)$



````python
# Funciones de las tres maneras diferentes de la función anterior 
# Funcion 1
def f1(x, r):
  """
  Función para calcular la población en un tiempo
  input:
    x: Población
    r: tasa de crecimiento
  output
    x: población final  
  """
  x = (r*x) - (r*(x)**2)
  return x

# Funcion 2
def f2(x, r):
  """
  Función para calcular la población en un tiempo
  input:
    x: Población
    r: tasa de crecimiento
  output
    x: población final  
  """
  x = (r)*(x - (x)**2)
  return x

# Funcion 3
def f3(x, r):
  """
  Función para calcular la población en un tiempo
  input:
    x: Población
    r: tasa de crecimiento
  output
    x: población final  
  """
  x = ((x)**2)*(((r)/(x))-r)
  return x

````

````python
# Mapas logisticos para las tres funciones
# Para la función 1
def mapa_logistico1(xi,r,iter_N):
  """
  Construye un array con la poblacion para un numero de tiempor definidos
  input:
    x: Población
    r: tasa de crecimiento
  output
    poblacion_:array con la poblacion final
  """
  poblacion1=np.zeros(iter_N)
  for i in range(0, iter_N):
    xn=f1(xi, r)
    xi=xn
    poblacion1[i]=xn

  return poblacion1  

# Para la función 2
def mapa_logistico2(xi,r,iter_N):
  """
  Construye un array con la poblacion para un numero de tiempor definidos
  input:
    x: Población
    r: tasa de crecimiento
  output
    poblacion_:array con la poblacion final
  """
  poblacion2=np.zeros(iter_N)
  for i in range(0, iter_N):
    xn=f2(xi, r)
    xi=xn
    poblacion2[i]=xn

  return poblacion2 

# Para la función 3
def mapa_logistico3(xi,r,iter_N):
  """
  Construye un array con la poblacion para un numero de tiempor definidos
  input:
    x: Población
    r: tasa de crecimiento
  output
    poblacion_:array con la poblacion final
  """
  poblacion3=np.zeros(iter_N)
  for i in range(0, iter_N):
    xn=f3(xi, r)
    xi=xn
    poblacion3[i]=xn

  return poblacion3   
````

````python
# Gráfica del primer mapa logistico

iter_N = 100   # Cantidad de iteraciones
N = np.arange(0, iter_N, 1)  # Arreglo de las iteraciones para el eje X
r = [1.9,2.9,3.1,3.5,3.9]  # Valores de r (tasa de crecimiento sugeridos)
x0 = 0.6 #  velor de x0 (Población)

plt.figure()
for i in range(0, len(r) ):

  p1 = mapa_logistico1(x0, r[i], iter_N)
  plt.plot(N, p1, label=""+str(r[i])+"")
  
plt.xscale('log')  
plt.show()
````

````python
# Gráfica del segundo mapa logistico

iter_N = 100
N = np.arange(0, iter_N, 1)
r = [1.9,2.9,3.1,3.5,3.9]
x0 = 0.6

plt.figure()
for i in range(0, len(r) ):

  p1 = mapa_logistico2(x0, r[i], iter_N)
  plt.plot(N, p1, label=""+str(r[i])+"")

plt.xscale('log')  
plt.show()
````

````python
# Gráfica del tercer mapa logistico

iter_N = 100
N = np.arange(0, iter_N, 1)
r = [1.9,2.9,3.1,3.5,3.9]
x0 = 0.6

plt.figure()
for i in range(0, len(r) ):

  p1 = mapa_logistico3(x0, r[i], iter_N)
  plt.plot(N, p1, label=""+str(r[i])+"")
  
plt.xscale('log')
plt.show()
````
