## Cancelación sustractiva



**serieSn**. Un ejemplo de la cancelacion substractiva es en la suma de series alternantes. Considere la serie $\mathrm{S}_N$,  escrita  matematicamente de tres formas diferentes:

- 
\begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1} 
\end{equation}

- Sumando números pares e impares.

\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

- Combinando las series de una forma analítica

\begin{equation}
\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}
\end{equation}



Las tres sumas $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$ son matemáticamente iguales, pero los resultados  pueden tener resultados numéricos diferentes.

````python
#Importamos las librerias

import numpy as np
import math as mt
import matplotlib.pyplot as plt
````

````python
#Definimos las series

def S1(n):                                #Definimos la serie 1
  suma_S1 = 0                             
  for i in range(1,n):                    #Hacemos un for para determinar el n-esimo termino
    num = ((-1)**i)*i                     #Definimos el i-esimo numerador  
    den = (i+1)                           #Definimos el i-esimo denominador
    suma_S1 += num/den                    #Obtenemos el i-esimo termino de la serie
  return suma_S1                          

#Cómo la serie 2 está divida en 2 series, para hallar el n-esimo término 
#la divideremos en 2 series, siendo 1 la parte izquierda y 2 la derecha
#luego sumaremos el n-esimo termino de cada una

def S2(n):                                #Definimos la serie 2
  suma1 = 0                               
  suma2 = 0
  for i in range (1,n):                   #Hacemos un for para determinar el n-esimo termino
    num1 = -(2*i-1)                       #Definimos el i1-esimo numerador 
    den1 = (2*i)                          #Definimos el i1-esimo denominador
    num2 = 2 * i                          #Definimos el i2-esimo numerador
    den2 = 2 * i + 1                      #Definimos el i2-esimo denominador
    suma1 += num1/den1                    #Obtenemos el i1-esimo termino de la serie
    suma2 += num2/den2                    #Obtenemos el i2-esimo termino de la serie
  suma_S2 = suma1 + suma2                 #Sumamos el n-esimo termino de ambas series
  return suma_S2

def S3(n):                                #Definimos la serie 3
  suma_S3 = 0
  for i in range(1,n):                    #Hacemos un for para determinar el n-esimo termino 
    suma_S3 += 1 / (2*i*(2*i+1))          #Obtenemos el i-esimo termino de la serie
  return suma_S3
````

````python
# Estudiamos la convergencia de las Series

n = int(input("Ingrese hasta que dato desea calcular "))   #Numero de datos que se graficaran
Nesimo = np.arange(1,n,1)

#Serie 1
s1 = []                                                    #Creamos una lista para guardar los primeros n valores
for i in Nesimo:                                           #Hacemos un for para determinar los primeros n terminos
  val1 = S1(i)                                             #Guardamos el i-esimo termino 
  s1.append(val1)                                          #Agregamos el i-esimo termino a una lista

#Serie 2
s2 = []                                                    #Creamos una lista para guardar los primeros n valores
for i in Nesimo:                                           #Hacemos un for para determinar los primeros n terminos
  val2 = S2(i)                                             #Guardamos el i-esimo termino 
  s2.append(val2)                                          #Agregamos el i-esimo termino a una lista

#Serie 3
s3 = []                                                    #Creamos una lista para guardar los primeros n valores
for i in Nesimo:                                           #Hacemos un for para determinar los primeros n terminos
  val3 = S3(i)                                             #Guardamos el i-esimo termino 
  s3.append(val3)                                          #Agregamos el i-esimo termino a una lista


print("""Para ver un poco la convergencia o divergencia se las series
Miraremos sus primeros {} valores en un gráfico\n

--Tablas de las series--

""".format(n))

#Grafico de la serie 1

plt.figure()
plt.title("Serie 1")
plt.plot(Nesimo,s1)
plt.xlabel("n")
plt.ylabel("Valores tomados por la serie")
plt.show()
print("")


#Grafico de la serie 2

plt.figure()
plt.title("Serie 2")
plt.plot(Nesimo,s2)
plt.xlabel("n")
plt.ylabel("Valores tomados por la serie")
plt.show()
print("")


#Grafico de la serie 3

plt.figure()
plt.title("Serie 3")
plt.plot(Nesimo,s3)
plt.xlabel("n")
plt.ylabel("Valores tomados por la serie")
plt.show()
````

## **Errores de redondeo**. Caos numérico en un mapa logístico y errores de punto flotante (Errores de redondeo)

> Un ejemplo clásico de caos,  es el comportamiento no lineal en las interaciones de un mapa logistico 

\begin{equation}
x_{n+1}=f(x_n)=rx_n(1-x_n)
\end{equation}

> con $x\in (0,1)$ y $r\in(0,4)$ se pueden producir varios comportamientos sorprendentes.


> ### Problema: 
Encontrar tres formas diferentes de expresar $f(x)$  y calcular la evolución de la misma condición inicial después de cientos de iteraciones. Para este problema, será extremadamente útil  ver sus resultados gráficamente; construya listas de números y llame la libreria matplotlib. 


>#### Hint: valores numéricos de r y x0 que puede tomar:


```python
    >>> r = [1.9, 2.9, 3.1, 3.5, 3.9]

    >>> x0 = 0.6 # any number in [0,1] will do here

    >>> numpoints = 100
       ```
       

````python
# Punto 2

x0 = 0.2
r  = 3.9
tiempos = 30 
N = np.arange(0, tiempos, 1)

def f1(x, r):#función de población 1
  x = r * x**2 * (1/x - 1)
  return x

def f2(x, r):#función de población 2
  x = r * x - r * x**2
  return x

def f3(x, r):#función de población 3
  x = np.sqrt((r**2)*(x**2)*(1-x)**2)
  return x

def mapa_logistico(x0, r, tiempos,F):
  """
  Construye un array con la poblacion para un numero de tiempos definido.
    x0: Poblacion inicial  
    r: tasa de crecimiento.
  Salida
    poblacion: Array con la Poblacion final
    """
  f = F
  poblacion = np.zeros(tiempos)
  for i in range(0, tiempos):
    xn = f(x0, r)
    x0 = xn
    poblacion[i] = xn
  return poblacion

#Mapa logistico de la Función 1
plt.figure(figsize=(20,5))
p1 = mapa_logistico(x0, r, tiempos,f1)
plt.subplot(131)
plt.plot(N, p1, color = 'red', label = '$rx^2(1/x - 1)$') 
plt.legend()
plt.show()

#Mapa logistico de la Función 2
plt.figure(figsize=(20,5))
p2 = mapa_logistico(x0, r, tiempos,f2)
plt.subplot(132)
plt.plot(N, p2,color = 'black', label = '$rx - rx^2$') 
plt.legend()
plt.show()

#Mapa logistico de la Función 3
plt.figure(figsize=(20,5))
p3 = mapa_logistico(x0, r, tiempos,f3)
plt.subplot(133)
plt.plot(N, p3,color = 'green',label = '$((r^2)*(x^2)*(1-x)^2)^{1/2}$') 
plt.legend()
plt.show()
````
