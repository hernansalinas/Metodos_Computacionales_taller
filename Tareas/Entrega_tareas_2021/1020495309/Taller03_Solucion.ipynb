```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# EN ESTA FUNCION ESTÁN IMPLEMENTADAS LAS 3 SERIES DIFERENTES, LA SERIE QUE SE DESEA CALCULAR SE ELIGE CON EL VALOR DE "x"
# Ejemplo: sumatoria(10,1), calcularia el decimo valor (N=10) de la primera serie (x=1)
def serie(N,x):
  #Error, por si el usuario ingresa un valor de x no establecido.
  suma_total = 0
  if x > 3:
    return "Error: debe ingresar un x entero en el siguiente intervalor; 0<x<4 "
  #Primera serie
  if x == 1: 
    i = 1
    suma_total = 0
    N = 2*N
    while i <= N: 
      ec_1 = i/(i+1)
      ec_2 = (-1)**i
      ec = ec_1 * ec_2
      suma_total = suma_total + ec
      i = i+1
    return suma_total
  #Segunda serie
  if x == 2:
    i = 1
    suma_total = 0
    suma_1 = 0
    suma_2 = 0
    while i <= N:
      ec_1 = -((2*i-1)/(2*i))
      ec_2 = ((2*i)/(2*i+1))
      suma_1 = suma_1 + ec_1
      suma_2 = suma_2 + ec_2
      i=i+1
    suma_total = suma_1 + suma_2
    return suma_total
  #Tercera serie
  if x == 3:
    i = 1
    suma_total = 0
    while i <= N:
      ec = 1/(2*i*(2*i+1))
      suma_total = suma_total + ec
      i = i+1
    return suma_total
```

```python
N = np.arange(1,500)
serie_1 = np.zeros(len(N))
serie_2 = np.zeros(len(N))
serie_3 = np.zeros(len(N))
for i,n in enumerate(N):
  serie_1[i] = serie(n,1)
  serie_2[i] = serie(n,2)
  serie_3[i] = serie(n,3)

plt.figure(figsize=(10,8))
plt.plot(N,serie_1, label="serie_1")
plt.plot(N,serie_2, label="serie_2")
plt.plot(N,serie_3, label="serie_3")
plt.xlabel("N")
plt.ylabel("SERIE")

plt.legend()
```

#En la grafica vemos que las series se parecen mucho para los mismo valores de N.
#La serie_1 que es alternante no se ve oscilar, y esto se debe a que cada valor N que se ingresa, esta serie se calcula hasta 2N.
#Si la serie alternante se calculara hasta N, la grafica sería la siguiente:

```python
#SERIE DE PRUEBA
def serie_prueba(N):
    i = 1
    suma_total = 0
    while i <= N: 
      ec_1 = i/(i+1)
      ec_2 = (-1)**i
      ec = ec_1 * ec_2
      suma_total = suma_total + ec
      i = i+1
    return suma_total
Q = np.arange(1,50)
serie_prueba_1 = np.zeros(len(Q))
for i,n in enumerate(Q):
  serie_prueba_1[i] = serie_prueba(n)
plt.figure(figsize=(10,8))
plt.plot(Q,serie_prueba_1, label="serie_prueba")
plt.xlabel("N")
plt.ylabel("SERIE")
plt.legend()

```

#Aquí vemos que para los valores de N pares, la serie de prueba tiende a el mismo valor que la serie_1 , pero para los valores impares tiende a un numero diferente. (por ello se calcula hasta 2N la serie_1)

```python
#AHORA CALCULEMOS LOS ERRORES
S_1 = ( serie_1 - serie_3 ) / serie_3
S_2 = ( serie_2 - serie_3 ) / serie_3 
plt.figure()
plt.plot(N,S_1, label="S_1")
plt.legend()
plt.figure()
plt.plot(N,S_2, label="S_2")
plt.legend()
```

```python

#Tomando la serie f(xn) = rxn(1-xn) rescribiendola de las siguientes formas

 #1. f(xn) = (r*xn) - (r*(xn)**2)
 #2. f(xn) = (r)*(xn - (xn)**2)
 #3. f(xn) = (r)*(xn - (xn)**2)

# Sabemos que:
#xn = población en un tiempo inicial
#r = diferencial de crecimiento con respecto al tiempo
#xf = poblacion en un tiempo final

# Primer función

def funcion1( xn , r ):

  xn = (r*xn) - (r*(xn)**2)
  
  return xn

#Segunda función

def funcion2( xn , r ):

  xf = (r)*(xn - (xn)**2)
  
  return xf

#Tercera función

def funcion3( xn , r ):

  xf = (r)*(xn - (xn)**2)
  
  return xf
```

```python
#Mapas logísticos

#xn = población en un tiempo inicial
#r = diferencial de crecimiento con respecto al tiempo
#xfn = poblacion en un tiempo final (con n = 1,2 o 3 dependiendo del numero del mapa logistico "mpln")

# Mapa logistico primera funcion

def mpl1(xn,r,N):

  xf1 = np.zeros(N)
  for i in range(0, N):
    xn = funcion1( xn, r )
    xf1[i] = xn

  return xf1

# Mapa logistico segunda funcion

def mpl2(xn,r,N):

  xf2 = np.zeros(N)
  for i in range(0, N):
    xn = funcion2( xn, r )
    xf2[i] = xn

  return xf2

# Mapa logistico tercera funcion

def mpl3(xn,r,N):

  xf3 = np.zeros(N)
  for i in range(0, N):
    xn = funcion3( xn, r )
    xf3[i] = xn

  return xf3
```

```python
# Graficacion
# Mapa logistico primera funcion

N = 100
array = np.arange(0,N, 1)
r = [1.9,2.9,3.1,3.5,3.9]
xinicial = 0.6

plt.figure()

for i in range (0, len(r)):

  plot = mpl1(xinicial, r[i], N)
  plt.plot(array, plot, label=""+str(r[i])+"")

plt.xscale("log")
plt.show()

# Graficacion
# Mapa logistico segunda funcion

N = 100
array = np.arange(0,N, 1)
r = [1.9,2.9,3.1,3.5,3.9]
xinicial = 0.6

plt.figure()

for i in range (0, len(r)):

  plot = mpl2(xinicial, r[i], N)
  plt.plot(array, plot, label=""+str(r[i])+"")

plt.xscale("log")
plt.show()

#graficacion
# Mapa logistico tercera funcion

N = 100
array = np.arange(0,N, 1)
r = [1.9,2.9,3.1,3.5,3.9]
xinicial = 0.6

plt.figure()

for i in range (0, len(r)):

  plot = mpl3(xinicial, r[i], N)
  plt.plot(array, plot, label=""+str(r[i])+"")

plt.xscale("log")
plt.show()
```
