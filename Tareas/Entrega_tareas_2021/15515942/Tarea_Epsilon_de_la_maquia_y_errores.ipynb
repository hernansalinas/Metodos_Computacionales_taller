```python
#Importamos librerías

import math as mt
import numpy as np
import matplotlib.pylab as plt 
```

```python
# Funciones para calcular por medio de series la exponencial negativa

def exponencial(x,N):

    suma = 0
  
    for n in range(0,N):

        suma  =  suma    +   (((-1)**n)*x**n)/mt.factorial(n)  #variable para acumular el valor de la serie
          
    return suma


def exponencial_1(x,N):

    suma1 = 0

    for n in range(0,N):

         suma1  =  suma1    +   (x**n)/mt.factorial(n)  #variable para acumular el valor de la serie
         
    return 1/suma1
```

```python
x = 10

print("El valor de la funcion nativa de python es : " ,  1/mt.exp(x))
print("El valor de la función alternante es  : " , exponencial(x,100))
print("El valor de la función sin alternar es  : " , exponencial(x,100))
```

```python
N_valores = np.arange( 3 , 50 , 1 )   #Arreglo de valores entre 3 y 50

#Se crean dos arreglos llenos de ceros con el tamaño del arreglo N_valores
exponencial_arreglo = np.zeros(len(N_valores))
exponencial1_arreglo = np.zeros(len(N_valores))

#Se llenan de valores los arreglos por medio de las funciones creadas anteriomente
for i ,n in enumerate(N_valores):
    exponencial_arreglo[i] = exponencial( 10 , n )
    exponencial1_arreglo[i] = exponencial_1( 10 , n )
    
```

```python
# Graficando los valores obtenidos para los arreglos 

plt.figure()
plt.plot(N_valores ,exponencial_arreglo , label = "SERIE ALTERNANTE exp (-10)")
plt.plot(N_valores ,exponencial1_arreglo , label = "SERIE 1 / exp (10)")

plt.legend() 

plt.show()
```

```python
# Funciones para calcular por medio de series la exponencial negativa

def serie(x,N):

    suma = 0
 
    for n in range(1,2*N):

        suma  =  suma    +   (((-1)**n)*n)/(n+1)  #variable para acumular el valor de la serie
          
    return suma


def serie_1(x,N):

    suma1 = 0
    suma2 = 0

    for n in range(1,N):

         suma1  =  suma1    +   (2*n-1)/(2*n)  #variable para acumular el valor de la serie
         suma2  =  suma2    +   (2*n)/(2*n+1)  #variable para acumular el valor de la serie
           
    return -suma1 + suma2



def serie_2(x,N):

    suma3 = 0
  
    for n in range(1,N):

        suma3  =  suma3    +   1/(2*n*(2*n+1))  #variable para acumular el valor de la serie
           
    return suma3
```

```python
N_valores = np.arange( 3 , 50 , 1 )  #Arreglo de valores entre 3 y 50

#Se crean tres arreglos llenos de ceros con el tamaño del arreglo N_valores
serie_arreglo = np.zeros(len(N_valores))
serie1_arreglo = np.zeros(len(N_valores))
serie2_arreglo = np.zeros(len(N_valores))

#Se llenan de valores los arreglos por medio de las funciones creadas anteriomente
for i ,n in enumerate(N_valores):
    serie_arreglo[i] = serie( 10 , n )
    serie1_arreglo[i] = serie_1( 10 , n )
    serie2_arreglo[i] = serie_2( 10 , n )
    
```

```python
# Graficando los valores obtenidos para los arreglos 

plt.figure()
plt.plot(N_valores ,serie_arreglo ,"r", label = "SERIE 1 (ALTERNANTE )")
plt.plot(N_valores ,serie1_arreglo ,"b", label = "SERIE 2 suma de dos series ")
plt.plot(N_valores ,serie2_arreglo ,"g", label = "SERIE 3")
plt.legend()
plt.show()
```

```python
# Funciones para calcular por medio de series la exponencial negativa

r= 2.9

def expresion(N):

    suma = 0.6
  
    for n in range(1,N):

        suma  =  suma    +   r*suma*(1-suma)  #variable para acumular el valor de la serie
    
    return suma


def expresion_1(N):

    suma1 = 0.6
    
    for n in range(1,N):

         suma1  =  suma1    +   r*(  suma1   - suma1*suma1)  #variable para acumular el valor de la serie
         
    return suma1 



def expresion_2(N):

    suma2 = 0.6
  
    for n in range(1,N):

        suma2  =  suma2    +   (suma2**2) * r *( (1/suma2) - 1 )  #variable para acumular el valor de la serie
    
    return suma2
```

```python
N_valores = np.arange( 1 , 100 , 1 )  #Creando un vector de valores de uno a 100

# A continuación se crean tres arreglos de ceros con la longitud de  N_valores
expresion_arreglo = np.zeros(len(N_valores)) 
expresion1_arreglo = np.zeros(len(N_valores))
expresion2_arreglo = np.zeros(len(N_valores))

# llenamos los arreglos con los valores que arrojan las funciones
for i ,n in enumerate(N_valores):
    expresion_arreglo[i] = expresion(  n )
    expresion1_arreglo[i] = expresion_1(  n )
    expresion2_arreglo[i] = expresion_2(  n )
    
```

```python
# Graficando los valores obtenidos para los arreglos 

plt.figure(figsize=(20,10))
plt.plot(N_valores ,expresion_arreglo ,"r",label = "Formula original")
plt.plot(N_valores ,expresion1_arreglo ,"b")
plt.plot(N_valores ,expresion2_arreglo ,"g")
plt.legend(fontsize = 22, loc = 1)
plt.show()
```
