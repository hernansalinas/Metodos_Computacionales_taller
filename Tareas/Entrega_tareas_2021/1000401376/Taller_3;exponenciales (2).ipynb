## Cancelación sustractiva



**serieSn**. Un ejemplo de la cancelacion substractiva es en la suma de series alternantes. Considere la serie $\mathrm{S}_N$,  escrita  matematicamente de tres formas diferentes:

- 
\begin{equation}
\mathrm{S}_N^{(1)} = \sum_{n = 1}^{2N} (-1)^n \frac{n}{n+1} 
\end{equation}

- Sumando números pares e impares.

\begin{equation}
\mathrm{S}_N^{(2)} = -\sum_{n = 1}^{N}\frac{2n-1}{2n} + \sum_{n=1}^{N} \frac{2n}{2n+1}
\end{equation}

- Combinando las series de una forma analítica

\begin{equation}
\mathrm{S}_N^{(3)} = \sum_{n = 1}^{N}\frac{1}{2n(2n+1)}
\end{equation}



Las tres sumas $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$ son matemáticamente iguales, pero los resultados  pueden tener resultados numéricos diferentes.


- Escribir un programa que realice el cálculo de $\mathrm{S}_N^{(1)}$, $\mathrm{S}_N^{(2)}$ y $\mathrm{S}_N^{(3)}$
- Estudiar la convergencia de la serie para los tres casos.

````
#Primera Serie
def serie_1(n):
  """
  Funcion para la primera serie, que recibe como entrada hasta que numero quiere llegar (n).
  y retornara la suma total, la variable encargada de esto es S_N_1
  """
  z = 2*n # Limite de la serie 
  S_N_1 = 0 
  for i in range(1, z+1):  # Vamos hasta z+1 porque el for va hasta z-1 y necesitamos ir hasta z
    cambio = ((-1)**i) #Ecuacion
    cte = i/(i+1)      #Ecuacion
    sumador = cambio * cte # La variable sumador es la que se encarga de utilizar la ecuacion (cte) 
    S_N_1 = S_N_1 + sumador #Le sumamos la variable sumador a S_N_1 para que esta lleve el conteo de la sumatoria
  return S_N_1

#Segunda Serie
def serie_2(n):
  """
  Funcion para la segunda serie, que recibe como entrada hasta que numero quiere llegar (n).
  y retornara la suma total, la variable encargada de esto es S_N_2
  """
  S_N_2 = 0

  """
  Esta serie esta compuesta por dos sumatorias, por esto decidí hace una funcion anidada, es decir hacer dos funciones para las dos sumatorias
  """
  def sumatoria1(n): #Definimos la primera sumatoria que recibe como argumento hasta donde quiere ir la sumatoria y devuelve la suma de la serie hasta n
    sumatoria_1 = 0 # Sumatoria_1 es la variable que devolvera la serie con la suma total de la primera sumatoria 
    for i in range(1, n+1): # Vamos hasta n+1 porque el for va hasta n-1 y necesitamos ir hasta n 
      cambio = (2*i-1)/(2*i) #Ecuacion 
      sumatoria_1 = sumatoria_1 + cambio #Le sumamos la variable cambio a sumatoria_1 para que esta lleve el conteo de la sumatoria
    return -sumatoria_1  

  def sumatoria2(n): #Definimos la segunda sumatoria que recibe como argumento hasta donde quiere ir la sumatoria y devuelve la suma de la serie hasta n
    sumatoria_2 = 0 #Sumatoria_2 es la variable que devolvera la serie con la suma total de la primera sumatoria 
    for i in range(1, n+1): #Vamos hasta n+1 porque el for va hasta n-1 y necesitamos ir hasta n
      cambio = 2*i/(2*i +1) #Ecuacion
      sumatoria_2 = sumatoria_2 + cambio #Le sumamos la variable cambio a sumatoria_2 para que esta lleve el conteo de la sumatoria
    return sumatoria_2

  """
  Como tenemos funciones anidadas es necesario llamarlas dentro de la propia funcion para poder que funcione, eso es loq ue hacemos a continuacion
  """
  cte_1 = sumatoria1(n) 
  cte_2 = sumatoria2(n)
  S_N_2 = cte_1+cte_2 #Sumamos el total de las dos sumatorias y se lo asignamos a la variable S_N_2, esta variable es la que devolvera la funcion principal
  return S_N_2
  
#Tercera Serie
def serie_3(n): 
  """
  Funcion para la tercera serie, que recibe como entrada hasta que numero quiere llegar (n).
  y retornara la suma total, la variable encargada de esto es S_N_3
  """
  S_N_3 = 0
  for i in range(1, n+1): #Vamos hasta n+1 porque el for va hasta n-1 y necesitamos ir hasta n
    z = 2*i #Ecuacion
    w = (z+1) #Ecuacion 
    cambio = 1/(z*w) #Ecuacin completa
    S_N_3 = S_N_3 + cambio #Le sumamos la variable cambio a S_N_3 para que esta lleve el conteo de la sumatoria
  return S_N_3
````

````
n = 10 #elegimos hasta donde queremos que vaya la serie 

metodo_s1 = serie_1(n) #LLmamos la primera funcion
metodo_s2 = serie_2(n) #LLmamos la segunda funcion
metodo_s3 = serie_3(n) #LLmamos la tercera funcion

print(f"la serie 1 con n = {n} es {metodo_s1}") #mostramos cuanto da la serie 1
print(f"la serie 2 con n = {n} es {metodo_s2}") #mostramos cuanto da la serie 2
print(f"la serie 3 con n = {n} es {metodo_s3}") #mostramos cuanto da la serie 3

difer_1 = metodo_s3 - metodo_s1
difer_2 = metodo_s2 - metodo_s1
difer_3 = metodo_s2 - metodo_s3

print(f"podemos ver que la diferencia entre la primera y tercera serie es {difer_1}, es decir un valor muy bajo")
print(f"podemos ver que la diferencia entre la segunda y primeraa serie es {difer_2}, es decir un valor muy bajo")
print(f"podemos ver que la diferencia entre la segunda y tercera serie es {difer_1}, es decir un valor muy bajo")
````

## **Errores de redondeo**. Caos numérico en un mapa logístico y errores de punto flotante (Errores de redondeo)

> Un ejemplo clásico de caos,  es el comportamiento no lineal en las interaciones de un mapa logistico 

\begin{equation}
x_{n+1}=f(x_n)=rx_n(1-x_n)
\end{equation}

> con $x\in (0,1)$ y $r\in(0,4)$ se pueden producir varios comportamientos sorprendentes.


> ### Problema: 
Encontrar tres formas diferentes de expresar $f(x)$  y calcular la evolución de la misma condición inicial después de cientos de iteraciones. Para este problema, será extremadamente útil  ver sus resultados gráficamente; construya listas de números y llame la libreria matplotlib. 


>#### Hint: valores numéricos de r y x0 que puede tomar:


```python
    >>> r = [1.9, 2.9, 3.1, 3.5, 3.9]

    >>> x0 = 0.6 # any number in [0,1] will do here

    >>> numpoints = 100
       ```
       

````
"""
Tomare las siguientes tres formas equivalentes de escribir la ecuacion; xn = x sub n

1) rxn(1-xn) Ecuacion 1
2) rxn-rxn^2 Ecuacion 2
3)xn^2((r/xn)-r) Ecuacion 3

"""
import matplotlib.pyplot as plt
import numpy as np 

def f1(x, r):
  """
  Funcion para calcular la poblacion en un tiempo.
  Recibe 
    x: Poblacion inicial  
    r: tasa de crecimiento.
  Devuelve 
  x: Poblacion final   
  """
  x = r*x*(1-x) #Ecuacion 1 para calcular la poblacion
  return x

def mapa_logistico1(x0, r, numpoints):
  """
  Esta funcion construye un array con la poblacion para un numero de tiempos definido (numpoints).
  Recibe
    x0: Poblacion inicial  
    r: tasa de crecimiento
    numpoints: Numero de iteraciones 
  Devuelve
    poblacion1: Array con la poblacion final
  """
  poblacion1 = np.zeros(numpoints)
  """
  Creamos un for para que vaya calculando la poblacion final con la funcion f, ademas de esto utilizamos 
  Poblacion1[i] = xn para remplazar los valores del array que es poblacion1
  """
  for i in range(0, numpoints): 
    xn = f1(x0, r) 
    x0 = xn
    poblacion1[i] = xn
  return poblacion1

def f2(x, r):
  """
  Funcion para calcular la poblacion en un tiempo.
  Recibe 
    x: Poblacion inicial  
    r: tasa de crecimiento.
  Devuelve 
  x: Poblacion final   
  """
  x = r*x-r*(x**2) #Ecuacion 2 para calcular la poblacion
  return x

def mapa_logistico2(x0, r, numpoints):
  """
  Esta funcion construye un array con la poblacion para un numero de tiempos definido (numpoints).
  Recibe
    x0: Poblacion inicial  
    r: tasa de crecimiento
    numpoints: Numero de iteraciones 
  Devuelve
    poblacion2: Array con la poblacion final
  """
  poblacion2 = np.zeros(numpoints)
  """
  Creamos un for para que vaya calculando la poblacion final con la funcion f2, ademas de esto utilizamos 
  Poblacion2[i] = xn para remplazar los valores del array que es poblacion2
  """
  for i in range(0, numpoints):
    xn = f2(x0, r)
    x0 = xn
    poblacion2[i] = xn
  return poblacion2

def f3(x, r):
  """
  Funcion para calcular la poblacion en un tiempo.
  Recibe 
    x: Poblacion inicial  
    r: tasa de crecimiento.
  Devuelve 
  x: Poblacion final   
  """
  x = (x**2)*(r/x-r) #Ecuacion 3 para calcular la poblacion 
  return x

def mapa_logistico3(x0, r, numpoints):
  """
  Esta funcion construye un array con la poblacion para un numero de tiempos definido (numpoints).
  Recibe
    x0: Poblacion inicial  
    r: tasa de crecimiento
    numpoints: Numero de iteraciones 
  Devuelve
    poblacion3: Array con la poblacion final
  """
  poblacion3 = np.zeros(numpoints)
  """
  Creamos un for para que vaya calculando la poblacion final con la funcion f3, ademas de esto utilizamos 
  Poblacion2[i] = xn para remplazar los valores del array que es poblacion3
  """
  for i in range(0, numpoints):
    xn = f3(x0, r)
    x0 = xn
    poblacion3[i] = xn
  return poblacion3

"""
Condiciones iniciales:
r: Diferentes tasas de crecimiento
x0: Poblacion inicial
numpoints =  numero de iteraciones
N = array de 0 a numpoints con salto de 1 en 1 
"""
r = [1.9, 2.9, 3.1, 3.5, 3.9]
x0 = 0.6
numpoints = 100
N = np.arange(0,numpoints,1)

"""
Creamos 3 graficas para cada ecuacion 
"""
plt.figure()
for i in range(0, len(r) ):

  p1 = mapa_logistico1(x0, r[i], numpoints) #utilizamos la funcion mapa_logistico1
  plt.plot(N, p1, label=""+str(r[i])+"")
plt.title("Ecuacion 1: x = r*x*(1-x)")

plt.figure()
for i in range(0, len(r) ):

  p2 = mapa_logistico2(x0, r[i], numpoints) #utilizamos la funcion mapa_logistico2
  plt.plot(N, p2, label=""+str(r[i])+"")
plt.title("Ecuacion 2: x = r*x-r*(x**2)")


plt.figure()
for i in range(0, len(r) ):

  p3 = mapa_logistico3(x0, r[i], numpoints) #utilizamos la funcion mapa_logistico3
  plt.plot(N, p3, label=""+str(r[i])+"")
plt.title("Ecuacion 3: x = (x**2)*(r/x-r)")


plt.show()
print("Podemos ver como despues de la 65 iteracion la grafica 3 varia respecto a la 1 y 2.Despues de la 80 iteracion las 3 graficas varian")
````
